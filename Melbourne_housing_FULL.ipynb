```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
import joblib
```

```python
data =  pd.read_csv(r'C:\Users\Dilpat Rai\Desktop\Pandas\Melbourne_housing_FULL.csv') 
```

```python
data.head(n = 5)
  
```

```python
data.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
  
```

```python
data.head(5)  
```

```python
features_df = pd.get_dummies(data, columns=['Suburb', 'CouncilArea', 'Type'])
```

```python
features_df
```

```python
del features_df["Price"] 
```

```python
X = features_df.to_numpy()
y = data['Price'].to_numpy() 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## We can change the max_depth for accuracy according to model fitting and we can also change the n_estimators according to results    

```python
model = ensemble.GradientBoostingRegressor(
n_estimators=150, #how many decision tree to build
learning_rate=0.1, #controls the rate at which additional decision trees influence the overall prediction.
max_depth=5, #(30) max number of layers(depth) 
min_samples_split=4, #min number of samples require to create new branch(binary) 
min_samples_leaf=6, #min no of samples appear in child node 
max_features=0.6, #total number of features presented to the model when etermining the best split
loss='huber' # model error rate
)
```

```python
model.fit(X_train, y_train) 
```

```python
joblib.dump(model, 'house_trained_model.pkl') 
```

```python
mse = mean_absolute_error(y_train, model.predict(X_train))
print ("Training Set Mean Absolute Error: %.2f" % mse)
```

```python
mse = mean_absolute_error(y_test, model.predict(X_test))
print ("Test Set Mean Absolute Error: %.2f" % mse)
```
